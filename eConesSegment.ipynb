```python
from scipy.spatial.distance import cdist
from operator import itemgetter, attrgetter


def closest_node(node, nodes):
    return nodes[cdist([node], nodes).argmin()]
def closest_node_index(node, nodes):
    return cdist([node], nodes).argmin()
def closest2_node_index(node, nodes):
    return (cdist([node], nodes).argsort()[0][0:2]).tolist()
def parse_int(s):
    try:
        res = int(eval(str(s)))
        if type(res) == int:
            return res
    except:
        return
```

```python
def E(q, r0, x, y):
   """Return the artificial field vector E=(Ex,Ey) due to cone q at r0."""
   den = ((x-r0[0])**2 + (y-r0[1])**2)*1.5
   return -q * (y - r0[1]) / den, q * (x - r0[0]) / den
```

```python
# find the mid point between two coordinates and the direction 
def midposdir (pos0,pos1):
    posx1 = pos1[0]
    posy1 = pos1[1]
    posx0 = pos0[0]
    posy0 = pos0[1]
    x = (posx1-posx0)/2.0
    y = (posy1-posy0)/2.0
    angle = np.arctan2(y, x) - np.pi/2
    
    return ((posx0+x,posy0+y), angle)
```

```python
# find the distance between two coordinates 
def distpos(pos0,pos1):
    posx1 = pos1[0]
    posy1 = pos1[1]
    posx0 = pos0[0]
    posy0 = pos0[1]
    return ((posx1-posx0)**2 + (posy1-posy0)**2)**.5
```

```python
# find distance from pos2 to line segemnt pos0,pos1
def distpostoline(pos0,pos1,pos2):
    x2 = pos2[0]
    y2 = pos2[1]
    x1 = pos1[0]
    y1 = pos1[1]
    x0 = pos0[0]
    y0 = pos0[1]
    return (((y1-y0)*x2)-((x1-x0)*y2)+(x1*y0)-(y1*x0))/((y1-y0)**2+(x1-x0)**2)**0.5
```

```python
def forcevectorpath (cones, entrypos, entrydir, entryspeed):
    #the first red cone has the first gate pair info
    redgate1cone = cones[0]
    print(redgate1cone)
    bluegate1cone = findconebynum(cones,redgate1cone[4])
    print(bluegate1cone)
    #the last red cone has the second gate pair info
    redcones = [(q,pos,te,conenum,paircone) for q, pos, te, conenum, paircone in cones if q == 1]
    redgate2cone = redcones[-1]
    print(redgate2cone)
    bluegate2cone = findconebynum(cones,redgate2cone[4])
    print(bluegate2cone)
    gate1midposdir = midposdir (redgate1cone[1],bluegate1cone[1])
    print("gate1midposdir",gate1midposdir)
    gate2midposdir = midposdir (redgate2cone[1],bluegate2cone[1])
    print(gate2midposdir)
    shortcut = distpos(gate1midposdir[0],gate2midposdir[0])
    print(shortcut)
    #speed = shortcut / 30.0
    #speed = 1.0
    speed = entryspeed
    #Xpos = gate1midposdir[0][0]
    #Ypos = gate1midposdir[0][1]
    Xpos = entrypos[0]
    Ypos = entrypos[1]
    Xposfinal = gate2midposdir[0][0]
    Yposfinal = gate2midposdir[0][1]
    print ("Start Xpos",Xpos)
    print ("Start Ypos",Ypos)
    #angle = gate1midposdir[1]
    angle = entrydir
    totaltries = 500
    tries = 0
    path = []
    disttoline = distpostoline(redgate2cone[1],bluegate2cone[1],(Xpos,Ypos)) 
    while (disttoline > 0 or distpos((Xpos,Ypos),gate2midposdir[0]) > 20) and tries < totaltries  :
        
        
    
        Ex = 0.0
        Ey = 0.0
        Ex = (gate2midposdir[0][0] - Xpos)/700.0
        Ey = (gate2midposdir[0][1] - Ypos)/700.0
        for conetype,coords,te,conenum,paircone in cones:
       
        
            ex, ey = E(conetype, coords, x=Xpos, y=Ypos)
            Ex += ex
            Ey += ey
            
            
            
            
        #print('angle',(angle*180/np.pi))
        oldangle = angle
        #print("Ex", Ex)
        #print("Ey", Ey)
        #print("distx", gate2midposdir[0][0] - Xpos)
        #print("disty", gate2midposdir[0][1] - Ypos)
        angleforce = np.arctan2(Ey, Ex)   
        angletotarget = np.arctan2( gate2midposdir[0][1] - Ypos,gate2midposdir[0][0] - Xpos )
        angle = angleforce
        if abs(angle - oldangle) < 0.1:
            speed = speed * 1.0
        else:
            speed = speed * 1.0
        
        newx=  np.cos(angle) * speed
        newy= np.sin(angle) * speed
        Xpos += newx
        Ypos += newy
        path.append((Xpos,Ypos))
        tries += 1
        disttoline = distpostoline(redgate2cone[1],bluegate2cone[1],(Xpos,Ypos))
        #print ('move',(newx, newy))
        #print(Xpos, Ypos)
        
        #print(tries)
        #print (disttoline)
        exitpos = (Xpos,Ypos)
        exitdir = angle
        exitspeed = speed
    return path, exitpos, exitdir, exitspeed
```

```python
def plotconeswithpath(cones,path):
    fig = plt.figure()
    ax = fig.add_subplot(111)

    # Add filled circles for the cones themselves
    cone_colors = {True: '#aa0000', False: '#0000aa'}
    for q, pos, te, conenum, paircone in cones:
        ax.add_artist(Circle(pos, 1.0, color=cone_colors[q>0]))
    
    for pos in path:
        ax.add_artist(Circle(pos, 0.5, color='orange'))
    
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_xlim(0,540)
    ax.set_ylim(0,600)
    ax.set_aspect('equal')
    ax.imshow(img, extent=[63, 474, 0, 721])
    plt.rc('figure', figsize= ( 80.0, 80.0))
    plt.show()
```
